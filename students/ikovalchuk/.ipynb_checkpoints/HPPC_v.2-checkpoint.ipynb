```python
def HppcResist(df1):
    
    cycles_num=np.unique(df1['cycle number'])
    Voltage=[]
    Current=[]
    capacity_ah=[]
    Resistance= [] 
    SOC = []

    for i in range (df1.shape[0]-1):
        curr_time= df1['time/s'].iloc[i]
        next_time= df1['time/s'].iloc[i+1]
        delta_time= next_time-curr_time
        capacity_ah.append(df1['I/mA'][i]*delta_time)
    
    capacity_ah= np.asarray(capacity_ah)
 #   print(sum(capacity_ah))

    for f in cycles_num:
        
        i=np.where(df1['cycle number']==f)
        current_df = df1[df1['cycle number'] == f]  
        for d in i[0]:
            if df1.loc[d,'I/mA'] < -0.000001 and df1.loc[d,'I/mA'] > -0.2:
                Voltage.append(df1['Ecell/V'].iloc[d])
                Current.append(df1['I/mA'].iloc[d])
                
## I checked that my code gets exactly what I need. I get the first and last value of my dataframe and divide by the current
## Question: when we apply negative current the SOC changes so what point to take (first/last)?       
                curent=capacity_ah[0: d]  
            
        #print(sum(curent)/sum(capacity_ah))
        
        SOC.append(sum(curent)/sum(capacity_ah))
        
        x=Voltage[-1]
        y=Voltage[0]
        z=Current[0]
        Resistance.append((x-y)/z)
        Voltage.clear()
        Current.clear()
        
    return (Resistance, SOC, capacity_ah)
## code cuts dataframe in cycles for each cycle finds where current was applied, Stores current and voltage during 
## these occurences and gives back the array final resistances for each cycle 

    
```

```python
def parse_hppc_data(df):
    """
    Process a coin cell HPPC file.
    
    Assumptions:
    ---------
      HPPC pulses are < -0.3 mA
      HPPC pulses are discharge pulses
    
    (These assumptions are true for cathodes and full cells.
    For anodes, the pulses may be on charge, not discharge.)
    
    Parameters:
    ---------
       df: an input Pandas DataFrame
       
    Returns:
    ---------
       cap_array: array of capacities in mAh
       soc_array: array SOCs (0-100%)
       res_array: array of resistances in Ohms
       
    """

    # Get unique cycle numbers
    cyc_num_array = df['cycle number'].unique()[0:-1]

    # Initialize the arrays
    cap_array = []
    res_array = []

    # Loop over each cycle number
    for cyc_num in cyc_num_array:

        curr_df = df[df['cycle number'] == cyc_num]

        ## Process capacities
        # Add the incremental capacity for this cycle number
        cap_array.append(np.max(curr_df['Q discharge/mA.h']))

        ## Process resistances
        # Pull out the discharge pulse data
        pulse_condition = curr_df['I/mA'] < -0.3
        pulse_df = curr_df[pulse_condition]

        # Find the index corresponding to V0
        idx_v0 = np.where(curr_df[pulse_condition])[0][0] - 1

        # Compute metrics
        V0 = curr_df['Ecell/V'].iloc[idx_v0] 
        V1 = pulse_df['Ecell/V'].iloc[-1]
        I = np.abs(np.mean(pulse_df['I/mA']))/1000

        res_array.append( (V0-V1)/I )

    # Take the cumulative summation of the capacity
    cap_array = np.cumsum(cap_array)
    soc_array = cap_array/np.max(cap_array) * 100
    
    return (cap_array, soc_array, res_array)
```

```python
for file in file_paths[0]:

    df = load_data(file)

    (cap_array, soc_array, res_array) = parse_hppc_data(df)
    
    plt.figure()
    plt.plot(soc_array, res_array)
    plt.title(file)
    
```
